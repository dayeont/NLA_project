```python
def rand_svd(A, S):
    Q, R = np.linalg.qr(A @ S)
    B = Q.T @ array_solution
    # print(B.shape)# the main goal is to do svd for small matrix
    U, sigma, V = np.linalg.svd(B)
    U = Q @ U
    return U, sigma, V
```

```python
directories = ['20', '40', '60', '80', '100', '120', '140']
time_dir_classic = []
time_dir_rand = []
error_rand = []
error_classic = []
array_shapes = []

for directory in directories:
    filenames = ['./' + directory + '/membrane_deflection_' + str(i) + '000000.vtu' for i in range(100)]
    array_solution = []
    reader = vtk.vtkXMLUnstructuredGridReader()
    
    for file_name in filenames:
        reader.SetFileName(file_name)
        reader.Update()  # Needed because of GetScalarRange
        data = reader.GetOutput()
        solution = VN.vtk_to_numpy(data.GetPointData().GetArray('w'))
        array_solution.append(solution)
    
    array_solution = np.array(array_solution).T
        
    n = array_solution.shape[0] 
    m = array_solution.shape[1]
    array_shapes.append(n)
    
    U_classic, s_classic, V_classic = np.linalg.svd(array_solution, full_matrices=False)    
    S = np.random.normal(np.random.rand() + 5, 100*np.random.rand(), (m, m))
    U, sigma, V = rand_svd(array_solution, S)
    
    time_normal_svd = %timeit -q -o np.linalg.svd(array_solution)
    time_rand_svd = %timeit -q -o rand_svd(array_solution, S)
    
    time_dir_classic.append(time_normal_svd.best)
    time_dir_rand.append(time_rand_svd.best)

    array_reconst_classic = U_classic @ np.diag(s_classic) @ V_classic
    norm_class = np.linalg.norm(array_solution - array_reconst_classic, 2)
    error_classic.append(norm_class)

    array_reconst_rand = U @ np.diag(sigma) @ V
    norm_rand = np.linalg.norm(array_solution - array_reconst_rand, 2)
    error_rand.append(norm_rand)
```

```python
plt.figure(figsize = (9, 8))
plt.rcParams.update({'font.size': 18})
plt.plot(array_shapes, error_classic, label='classic')
plt.plot(array_shapes, error_rand, label='random')
plt.title("Errors vs number of rows for different SVD algorithms\n")
plt.ylabel("Error(2 norm of matrix difference)")
plt.xlabel("Number of rows")
plt.legend(loc='best')
plt.grid('on')
plt.savefig("errors_vs_rows.png")
plt.show()
```
